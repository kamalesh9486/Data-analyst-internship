```python
import bs4
```

```python
import pandas as pd
```

```python
from bs4 import BeautifulSoup
```

```python
url='https://en.wikipedia.org/wiki/List_of_most-played_mobile_games_by_player_count'
```

```python
ws=requests.get(url)
```

```python
soup = BeautifulSoup(ws.text,'html')
```

```python
table = soup.find('table', class_= 'wikitable')
```

```python
table.find_all('th')
```

```python
headers = []
for th in table.find_all('th'):
    headers.append(th.text.strip())
```

```python
rows = []
for tr in table.find_all('tr')[1:]:
    #print(tr)
    cells = tr.find_all(['td', 'th'])
    #print(cells)
    row = [cell.text.strip() for cell in cells]
   # print(row)
    rows.append(row)
print(rows)
```

```python
df = pd.DataFrame(rows, columns=headers)
```

```python
df
```

```python
df1=df[1:]
```

```python
df1
```

```python
import re
from io import StringIO
```

```python
df1['Game'] = df1['Game'].str.strip().str.replace('"', '', regex=False)
```

```python
df1['As of'] = pd.to_datetime(df1['As of'], errors='coerce')
```

```python
def clean_player_count(count_str):
    # Check if the input is a string
    if isinstance(count_str, str):
        # Use regular expression to find all numeric values in the string
        numbers = re.findall(r'\d+', count_str)

        # Join all found numbers and convert to integer
        if numbers:
            return int(''.join(numbers))
```

```python
df['Player count[a]'] = df['Player count[a]'].apply(clean_player_count)
```

```python
df1['Release date'] = pd.to_datetime(df1['Release date'], errors='coerce')
```

```python
df1['Publisher(s)'] = df1['Publisher(s)'].str.strip()


```
